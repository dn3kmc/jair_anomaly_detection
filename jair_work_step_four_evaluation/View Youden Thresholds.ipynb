```python
import sys  
sys.path.append("../time_series")  
sys.path.append("../anomaly_detection_methods")

import anomaly_detection_methods_helpers as ah
from time_series import TimeSeries
from os import listdir
from os.path import isfile, join
import joblib
import pandas as pd
pd.options.display.float_format = '{:,.4f}'.format

import numpy as np
from sklearn.metrics import roc_curve, auc
import math
import matplotlib.pyplot as plt
```

```python
dataset_column = []
anomalous_column = []
fb_column = []
glim_column  = []
hstree_column = []
htm_column = []
matrix_profile_column = []
pbad_column = []
sarimax_column = []
stl_column = []
twitter_column = []
vae_column = []
windowed_gaussian_column = []

mypath = "../jair_work_step_one_determine_characteristics/"
for f in listdir(mypath):
    if "ts_object" in f:
        ts = joblib.load(mypath + f)
        print(ts.name)
        dataset_column.append(ts.name[:-4])
        threshold_dict = joblib.load("dataset_youden_thresholds/" + ts.name + "_youden_thresholds")
        # print(threshold_dict)
        
        if "Anomalous" in threshold_dict:
            anomalous_column.append(threshold_dict["Anomalous"])
        else:
            anomalous_column.append("N/A")

        if "Prophet" in threshold_dict:
            fb_column.append(threshold_dict["Prophet"])
        else:
            fb_column.append("N/A")
            
        if "GLiM" in threshold_dict:
            glim_column.append(threshold_dict["GLiM"])
        else:
            glim_column.append("N/A")
        
        if "HS Tree" in threshold_dict:
            hstree_column.append(threshold_dict["HS Tree"])
        else:
            hstree_column.append("N/A")
        
        if "HTM" in threshold_dict:
            htm_column.append(threshold_dict["HTM"])
        else:
            htm_column.append("N/A")
        
        if "Matrix Profile" in threshold_dict:
            matrix_profile_column.append(threshold_dict["Matrix Profile"])
        else:
            matrix_profile_column.append("N/A")
        
        if "PBAD" in threshold_dict:
            pbad_column.append(threshold_dict["PBAD"])
        else:
            pbad_column.append("N/A")
        
        if "SARIMAX" in threshold_dict:
            sarimax_column.append(threshold_dict["SARIMAX"])
        else:
            sarimax_column.append("N/A")
        
        if "STL" in threshold_dict:
            stl_column.append(threshold_dict["STL"])
        else:
            stl_column.append("N/A")
        
        if "Twitter" in threshold_dict:
            twitter_column.append(threshold_dict["Twitter"])
        else:
            twitter_column.append("N/A")
        
        if "VAE" in threshold_dict:
            vae_column.append(threshold_dict["VAE"])
        else:
            vae_column.append("N/A")
        
        if "Windowed Gaussian" in threshold_dict:
            windowed_gaussian_column.append(threshold_dict["Windowed Gaussian"])
        else:
            windowed_gaussian_column.append("N/A")
```

```python
youden_threshold_df = pd.DataFrame({"Dataset": dataset_column,
                                    "Anomalous": anomalous_column,
                                    "Prophet": fb_column,
                                    "GLiM": glim_column,
                                    "HS Tree": hstree_column,
                                    "HTM": htm_column,
                                    "Matrix Profile": matrix_profile_column,
                                    "PBAD": pbad_column,
                                    "SARIMAX": sarimax_column,
                                    "STL": stl_column,
                                    "Twitter": twitter_column,
                                    "VAE": vae_column,
                                    "Windowed Gaussian": windowed_gaussian_column})
```
